```python
import re
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sample_code = """

def divide(a,b):
    return a/b

"""
```

```python
issues = []

if "def " in sample_code:

    if "->" not in sample_code:
        issues.append(
            "Missing type hints"
        )

if '"""' not in sample_code:
    issues.append(
        "Missing docstring"
    )

if "/b" in sample_code:

    if "b==0" not in sample_code:
        issues.append(
            "Possible division by zero"
        )

issues
```

```python
score = 100 - len(issues)*15

print("Code Quality Score:", score)

print()

for i,issue in enumerate(issues,1):

    print(f"{i}. {issue}")
```

```python
fixed_code = """

def divide(a:int,b:int)->float:
    \"\"\"Safely divide two numbers\"\"\"

    if b == 0:
        raise ValueError(
            "Division by zero"
        )

    return a / b

"""

print(fixed_code)
```

```python
plt.figure(figsize=(6,4))

plt.bar(
    ["Quality Score"],
    [score]
)

plt.ylim(0,100)

plt.title(
    "Code Quality Assessment"
)

plt.show()
```
